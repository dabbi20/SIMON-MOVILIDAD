```python
# ========================
# 1. Importar librer√≠as
# ========================
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# ========================
# 2. Subir archivo .parquet
# ========================
print("üìÇ Seleccion√° tu archivo .parquet")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
print(f"‚úîÔ∏è Archivo cargado: {file_name}")
```

```python
# ========================
# 3. Leer Parquet
# ========================
df = pd.read_parquet(file_name)

print("üìù Vista previa del DataFrame original:")
display(df.head())

print("\nüìò Schema estilo Spark (printSchema):")
for col, dtype in df.dtypes.items():
    print(f"{col}: {dtype}")
```

```python
# ===========================================================
# 4. GENERAR LAT/LONG DUMMY REALISTAS PARA TODOS LOS VEH√çCULOS
# ===========================================================
# Obtenemos todos los IDs √∫nicos del dataset
unique_ids = df["PULocationID"].unique()

# Generamos coordenadas aleatorias dentro del bounding box real de NYC
coords_df = pd.DataFrame({
    "PULocationID": unique_ids,
    "pickup_latitude": np.random.uniform(40.55, 40.95, size=len(unique_ids)),
    "pickup_longitude": np.random.uniform(-74.15, -73.70, size=len(unique_ids))
})

# Unimos coordenadas al df
df = df.merge(coords_df, on='PULocationID', how='left')

print("\nüìç Vista previa del DataFrame con lat/long agregadas:")
display(df.head())
```

```python
# ===========================================================
# 5. SALIDA 1 ‚Äî √öltima ubicaci√≥n conocida por veh√≠culo
# ===========================================================

df['vehicle_id'] = df['PULocationID']

# Ordenamos por timestamp
df_sorted = df.sort_values(by='tpep_pickup_datetime', ascending=False)

# √öltimo registro por veh√≠culo
df_last = df_sorted.groupby('vehicle_id').head(1)[[
    'vehicle_id',
    'tpep_pickup_datetime',
    'pickup_latitude',
    'pickup_longitude'
]]

df_last = df_last.rename(columns={
    'tpep_pickup_datetime': 'ultimo_timestamp',
    'pickup_latitude': 'ultima_latitud',
    'pickup_longitude': 'ultima_longitud'
})

print("\nüöñ Resultado final ‚Äì √öltima ubicaci√≥n por veh√≠culo:")
display(df_last)
```

```python
# Guardamos CSV
output_name = "ultima_ubicacion.csv"
df_last.to_csv(output_name, index=False)
files.download(output_name)
print(f"\nüíæ Archivo generado: {output_name}")
```

```python
# ===========================================================
# 6. SALIDA 2 ‚Äî Reporte de Viajes por Hora
# ===========================================================

# Extraer la hora del pick-up
df['hora_del_dia'] = df['tpep_pickup_datetime'].dt.hour

# Conteo por hora
df_hourly = df.groupby('hora_del_dia').size().reset_index(name='total_viajes')

# Ordenamos por hora 0..23
df_hourly = df_hourly.sort_values(by='hora_del_dia')

print("\n‚è∞ Reporte de viajes por hora del d√≠a:")
display(df_hourly)
```

```python
# Guardamos CSV
output_hour = "viajes_por_hora.csv"
df_hourly.to_csv(output_hour, index=False)
files.download(output_hour)
print(f"\nüíæ Archivo generado: {output_hour}")
```
